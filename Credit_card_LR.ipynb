```python
X_features = list( credit_df.columns )
X_features.remove( 'status' )
X_features
```

```python
encoded_credit_df = pd.get_dummies( credit_df[X_features],
drop_first = True,
dtype=int )
```

```python
import statsmodels.api as sm
Y = credit_df.status
X = sm.add_constant( encoded_credit_df )
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
Y,
test_size = 0.3,
random_state = 42)
```

```python
import statsmodels.api as sm
logit = sm.Logit(y_train, X_train)
logit_model = logit.fit()
```

```python
logit_model.summary2()
```

```python
def get_significant_vars(lm):
    var_p_vals_df = pd.DataFrame(lm.pvalues)
    var_p_vals_df['vars'] = var_p_vals_df.index
    var_p_vals_df.columns = ['pvals', 'vars']
    return list(var_p_vals_df[var_p_vals_df.pvals <= 0.05]['vars'])
```

```python
significant_vars = get_significant_vars( logit_model )
significant_vars
```

```python
final_logit = sm.Logit( y_train,

sm.add_constant( X_train[significant_vars] ) ).fit()
```

```python
final_logit.summary2()
```

```python
y_pred_df = pd.DataFrame( { "actual": y_test,

"predicted_prob": final_logit.predict(
sm.add_constant( X_test[significant_vars] ) ) } )
```

```python
y_pred_df.sample(10, random_state = 42)
```

```python
y_pred_df['predicted'] = y_pred_df.predicted_prob.map(
lambda x: 1 if x > 0.5 else 0)

y_pred_df.sample(10, random_state = 42)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn import metrics

def draw_cm(actual, predicted):
    # Create confusion matrix
    cm = metrics.confusion_matrix(actual, predicted, labels=[1, 0])

    sn.heatmap(cm, annot=True, fmt='.2f',
               xticklabels=["Bad credit", "Good Credit"],
               yticklabels=["Bad credit", "Good Credit"],
               cmap='Blues')

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics

def draw_roc(actual, probs):
    fpr, tpr, thresholds = metrics.roc_curve(actual, probs, drop_intermediate=False)
    auc_score = metrics.roc_auc_score(actual, probs)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)' % auc_score, color='blue')
    plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

    return fpr, tpr, thresholds
```

```python
fpr, tpr, thresholds = draw_roc( y_pred_df.actual,y_pred_df.predicted_prob)
```

```python
from sklearn import metrics
import numpy as np

def get_total_cost(actual, predicted, cost_FP, cost_FN):
    cm = metrics.confusion_matrix(actual, predicted, labels=[1, 0])
    TP = cm[0, 0]
    FN = cm[0, 1]
    FP = cm[1, 0]
    TN = cm[1, 1]

    total_cost = FP * cost_FP + FN * cost_FN
    return total_cost
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample: y_pred_df should have 'actual' and 'predicted_prob' columns
# Replace this with your actual prediction dataframe
y_pred_df = pd.DataFrame({
    'actual': np.random.randint(0, 2, 1000),
    'predicted_prob': np.random.rand(1000)
})

def gain_lift_chart(y_true, y_prob, n_bins=10):
    df = pd.DataFrame({'actual': y_true, 'predicted_prob': y_prob})
    df = df.sort_values(by='predicted_prob', ascending=False).reset_index(drop=True)
    df['decile'] = pd.qcut(df.index, q=n_bins, labels=False)

    grouped = df.groupby('decile').agg(
        total=('actual', 'count'),
        positives=('actual', 'sum')
    ).sort_index(ascending=False).reset_index()

    grouped['cumulative_positives'] = grouped['positives'].cumsum()
    grouped['cumulative_total'] = grouped['total'].cumsum()

    total_positives = df['actual'].sum()
    total = len(df)

    grouped['gain'] = grouped['cumulative_positives'] / total_positives
    grouped['lift'] = grouped['gain'] / (grouped['cumulative_total'] / total)

    return grouped

# Run function
gain_lift_df = gain_lift_chart(y_pred_df['actual'], y_pred_df['predicted_prob'])

# Plot charts
plt.figure(figsize=(12, 5))

# Gain Chart
plt.subplot(1, 2, 1)
plt.plot(gain_lift_df['gain'], marker='o', label='Gain')
plt.plot(np.linspace(0, 1, len(gain_lift_df)), linestyle='--', color='grey', label='Baseline')
plt.title('Gain Chart')
plt.xlabel('Deciles')
plt.ylabel('Cumulative Gain')
plt.legend()

# Lift Chart
plt.subplot(1, 2, 2)
plt.plot(gain_lift_df['lift'], marker='o', color='green', label='Lift')
plt.axhline(y=1, linestyle='--', color='grey', label='Baseline')
plt.title('Lift Chart')
plt.xlabel('Deciles')
plt.ylabel('Lift')
plt.legend()

plt.tight_layout()
plt.show()
```
